```python
import pandas as pd
star_wars = pd.read_csv("https://raw.githubusercontent.com/fivethirtyeight/data/master/star-wars-survey/StarWars.csv", encoding="ISO-8859-1")
print(star_wars.columns.values)
star_wars.head()
```

```python
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
print(star_wars.shape)
star_wars.head()
```

```python
print(star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].value_counts())
print(star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].value_counts())
```

```python
# convert Yes and No to Boolean
yes_no = {
    
    "Yes": True,
    "No": False
    
}

for col in ["Have you seen any of the 6 films in the Star Wars franchise?", "Do you consider yourself to be a fan of the Star Wars film franchise?" ]:
    star_wars[col] = star_wars[col].map(yes_no)
```

```python
import numpy as np

seen = {
    
    "Star Wars: Episode I  The Phantom Menace": True,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True,
    np.nan: False
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(seen)

star_wars.head()
```

```python
movie_titles = {
    
    "Which of the following Star Wars films have you seen? Please select all that apply." : "seen_1",
    "Unnamed: 4": "seen_2",
    "Unnamed: 5": "seen_3",
    "Unnamed: 6": "seen_4",
    "Unnamed: 7": "seen_5",
    "Unnamed: 8": "seen_6"
}

star_wars = star_wars.rename(columns=movie_titles)
star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
movie_ranking = {
    
    "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
    "Unnamed: 10": "ranking_2",
    "Unnamed: 11": "ranking_3",
    "Unnamed: 12": "ranking_4",
    "Unnamed: 13": "ranking_5",
    "Unnamed: 14": "ranking_6",
     
}

star_wars = star_wars.rename(columns=movie_ranking)
star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]].mean()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.bar(range(6), star_wars[star_wars.columns[9:15]].mean())
```

Star Wars 3,4 and 5 have the lowest ranking and are the most popular

```python
star_wars[star_wars.columns[3:9]].sum()
```

```python
plt.bar(range(6), star_wars[star_wars.columns[3:9]].sum())
```

The early Star Wars movies (3,4 and 5) have been seen by more fans. That reinforces the thought that they are more popular

```python
# split the data set in fan and not fan
fan = star_wars[star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] == True]
print(fan.shape)
not_fan = star_wars[star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] == False]
print(not_fan.shape)
```

```python
fan[fan.columns[9:15]].mean()
```

```python
plt.bar(range(6), fan[fan.columns[9:15]].mean())
```

Fans rate the earlier movies much better than the latest movies

```python
fan[fan.columns[3:9]].sum()
```

```python
plt.bar(range(6), fan[fan.columns[9:15]].sum())
```

The earlier movies are much more popular under fans

```python
not_fan[not_fan.columns[9:15]].mean()
```

```python
plt.bar(range(6), not_fan[not_fan.columns[9:15]].mean())
```

Non fans rate the latest and earlier movies quite evenly with 2 and 3 as the best

```python
not_fan[not_fan.columns[3:9]].sum()
```

```python
plt.bar(range(6), not_fan[not_fan.columns[3:9]].sum())
```

Under the non fans the earlier movies are still more popular
